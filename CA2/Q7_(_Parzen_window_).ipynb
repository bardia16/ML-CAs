```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

# Dataset

```python
N = 1000
np.random.seed(1)
X = np.concatenate((np.random.normal(0, 1, int(0.3 * N)),
                    np.random.normal(5, 1, int(0.7 * N))))[:, np.newaxis]
```

# Parzen window estiamtion

```python
def kernel_density_estimation(data, window_size, x_range):
    kde = []
    for x in x_range:
        kernel_values = norm.pdf((x - data) / window_size)
        kde.append(np.mean(kernel_values) / window_size)
    return np.array(kde)

window_sizes = [0.1, 0.5, 1, 10]

x_range = np.linspace(-5, 10, 1000)

plt.figure(figsize=(12, 8))

for window_size in window_sizes:
    kde_values = kernel_density_estimation(X, window_size, x_range)
    plt.plot(x_range, kde_values, label=f'Window Size = {window_size}')

plt.hist(X, bins=50, density=True, alpha=0.5, label='Histogram of Data')
plt.title('Kernel Density Estimation with Gaussian Kernel')
plt.xlabel('X')
plt.ylabel('Density')
plt.legend()
plt.show()
```
